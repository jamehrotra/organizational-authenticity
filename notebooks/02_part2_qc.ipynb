# Part 2 QC — Proxy Statement Coverage and Analysis

Inspect DEF 14A filing coverage and theme distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/part2_dataset.csv')
print(df.shape)
df.head()
```

```python
# Coverage by year
coverage = df.groupby('year')['has_filing'].sum().reset_index()
coverage.columns = ['year', 'filings_found']
coverage['total'] = df.groupby('year').size().values
coverage['pct'] = (coverage['filings_found'] / coverage['total'] * 100).round(1)
print(coverage)
```

```python
# Theme emphasis by sector
theme_cols = [c for c in df.columns if c.startswith('theme_') and c != 'theme_categories']
sector_themes = df.groupby('sector')[theme_cols].mean()

fig, ax = plt.subplots(figsize=(14, 6))
sns.heatmap(sector_themes.T, cmap='YlOrRd', vmin=0, vmax=3, annot=True, fmt='.1f', ax=ax)
ax.set_title('Mean Theme Scores by Sector (Proxy Statements)')
plt.tight_layout()
plt.show()
```

```python
# DEI and ESG emphasis over time
time_trends = df.groupby('year')[['dei_emphasis', 'esg_emphasis', 'employee_emphasis']].mean()
time_trends.plot(figsize=(10, 5), marker='o', title='DEI / ESG / Employee Emphasis in Proxy Statements Over Time')
plt.ylabel('Mean Emphasis Score (0-3)')
plt.grid(alpha=0.3)
plt.show()
```
